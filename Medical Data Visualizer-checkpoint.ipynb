```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Gokul\Downloads\medical_examination.csv')
```

```python
# Calculate BMI and add overweight column
df['BMI'] = df['weight'] / ((df['height'] / 100) ** 2)
df['Overweight'] = df['BMI'].apply (lambda x: 1 if x > 25 else 0)
```

```python
df['Overweight']
```

#Normalize the data by making 0 always good and 1 always bad for cholesterol and gluc columns.

```python
df['gluc'] = df['gluc'].apply(lambda x:0 if x == 1 else 1)
df['gluc']
```

```python
df['cholesterol'] = df['cholesterol'].apply(lambda x:0 if x == 1 else 1)
df['cholesterol']
```

```python
#Convert data into long format
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 'alco', 'active', 'smoke'])
df_cat
```

```python
# Create a catplot
cat_plot = sns.catplot(data=df_cat, kind='count', x='variable', hue='value', col='cardio')
cat_plot.set_axis_labels('variable', 'total')
cat_plot.set_titles('{col_name}')
cat_plot.savefig('catplot.png')
```

```python
#Filter out incorrect data
df = df[(df['ap_lo'] <= df['ap_hi']) &
        (df['height'] >= df['height'].quantile(0.025))&
        (df['height'] <= df['height'].quantile(0.975))&
        (df['weight'] >= df['weight'].quantile(0.025))&
        (df['weight'] <= df['weight'].quantile(0.975))]
df
```

```python
# Create correlation matrix
corr_matrix = df.corr()
```

```python
# Plot the heatmap
mask = np.triu(corr_matrix)
sns.heatmap(corr_matrix, annot=True, fmt='.1f', cmap='coolwarm',mask = mask)
plt.savefig('heatmap.png')
```
