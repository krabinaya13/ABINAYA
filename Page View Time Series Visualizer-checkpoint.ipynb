```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\Gokul\Downloads\fcc-forum-pageviews.csv' ,parse_dates=['date'], index_col='date')
data
```

```python
data.head()
```

```python
data = data[
    (data['value'] >= data['value'].quantile(0.025)) &
    (data['value'] <= data['value'].quantile(0.975))
]
```

```python
def draw_line_plot():
    # Step 3: Create a line chart
    plt.figure(figsize=(15, 5))
    plt.plot(data.index, data['value'], color='r')
    plt.title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
    plt.xlabel('Date')
    plt.ylabel('Page_views')
    plt.xticks(rotation=45)
    plt.show()
```

```python
def draw_bar_plot():
    data_bar = data.copy()
    data_bar['year'] = data_bar.index.year
    data_bar['month'] = data_bar.index.strftime('%B')
    data_bar = data_bar.groupby(['year', 'month'])['value'].mean().unstack()
    
    
    plt.figure(figsize=(15, 6))
    ax = data_bar.plot(kind='bar', legend=True)
    plt.legend(title='Months', labels=sorted(set(data_bar.columns), key=lambda x: list(data_bar.columns).index(x)))
    plt.xlabel('Years')
    plt.ylabel('Average Page Views')
    plt.title('Average daily Page Views (May 2016-December 2019)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
def draw_box_plot():
    data_box = data.copy()
    data_box.reset_index(inplace=True)
    data_box['year'] = [d.year for d in data_box['date']]
    data_box['month'] = [d.strftime('%b') for d in data_box['date']]

#Year-wise Box Plot (Trend)    
    plt.figure(figsize=(15,6))
    ax = sns.boxplot(x='year', y='value', data=data_box)
    ax.set_xlabel('Year')
    ax.set_ylabel('Page Views')
    ax.set_title('Year-wise Box Plot (Trend)')
    plt.show()
    
#Month-wise Box Plot (Seasonality)
    plt.figure(figsize=(15, 6))
    ax = sns.boxplot(x ='month', y ='value' ,data= data_box,order=[
                     'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    ax.set_xlabel('month')
    ax.set_ylabel('Page Views')
    ax.set_title('Month-wise Box Plot (Seasonality)')
    plt.show()
    
    
```

```python
line_plot = draw_line_plot()
bar_plot =draw_bar_plot()
box_plot =draw_box_plot()
```
